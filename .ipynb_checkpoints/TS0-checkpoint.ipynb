En primera instancia vamos a generar una señal senoidal , la cual va a estar compuesta por los siguientes parámetros:
- vmax = Amplitud máxima de la señal.
- dc = Offset, desplaza la señal para arriba o para abajo.
- ff = Frecuencia de la señal ( ciclos por segundo).
- ph = Fase.
- nn = Cantidad de muestras. 
- fm = Frecuencia de muestreo.

Genero la función que crea una señal senoidal.

```python
import numpy as np
import matplotlib.pyplot as plt

def mi_funcion_sen(vmax, dc, ff, ph, nn, fm):

    tt = np.arange(nn) / fm 
    xx = dc + vmax * np.sin(2 * np.pi * ff * tt + ph)

    return tt, xx
```

Llamo a la función y le asigno valores a sus parámetros, inicialmente la frecuencia de la señal va a ser de ff = 1 Hz. Luego grafico el resultado.

```python
vmax = 1    
dc = 0
ff = 1
ph = 0
nn = 100
fm = 100
tt, xx = mi_funcion_sen(vmax, dc, ff , ph, nn, fm)

plt.plot(tt,xx)
plt.title("Señal senoidal ff = 1 Hz")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid()
```

# <center>Bonus<center/>

Realizo modificaciones en la frecuencia de la señal (ff) para observar gráficamente su comportamiento ante estas alteraciones.
Los valores que voy a utilizar de ff son:
- ff1 = 500 Hz
- ff2 = 999 Hz
- ff3 = 1001 Hz
- ff4 = 2001 Hz

```python
#Primer Experimento

vmax = 1    
dc = 0
ff = 500
ph = 0
nn = 100
fm = 100

tt1, xx1 = mi_funcion_sen(vmax, dc, ff , ph, nn, fm)

#Segundo Experimento

vmax = 1    
dc = 0
ff = 999
ph = 0
nn = 100
fm = 100

tt2, xx2 = mi_funcion_sen(vmax, dc, ff , ph, nn, fm)

#Tercer Experimento

vmax = 1    
dc = 0
ff = 1001
ph = 0
nn = 100
fm = 100

tt3, xx3 = mi_funcion_sen(vmax, dc, ff , ph, nn, fm)

#Cuarto Experimento

vmax = 1    
dc = 0
ff = 2001
ph = 0
nn = 100
fm = 100

tt4, xx4 = mi_funcion_sen(vmax, dc, ff , ph, nn, fm)

#------Graficos------

#Subplots de experimentos

plt.figure()

plt.subplot(2,2,1)
plt.plot(tt1, xx1)
plt.title("Señal 1: ff = 500 Hz")
plt.grid()


plt.subplot(2,2,2)
plt.plot(tt2, xx2)
plt.title("Señal 2: ff = 999 Hz")
plt.grid()

plt.subplot(2,2,3)
plt.plot(tt3, xx3)
plt.title("Señal 3: ff = 1001 Hz")
plt.grid()

plt.subplot(2,2,4)
plt.plot(tt4, xx4)
plt.title("Señal 4: ff = 2001 Hz")
plt.grid()
plt.tight_layout() #Acomodo los espacios entre graficos
```

## Conclusión 

Por el teorema de Nyquist sabemos que para reconstruir una señal analogica
la frecuencia de muestreo debe ser al menos el doble de la frecuencia maxima presente en 
dicha señal para prevenir de manera teorica el aliasing (solapamiento). En nuestro caso sería pedir ff > 2fm. Entonces, los comportamientos observados se deben a que cuando la frecuencia de la señal supera la frecuencia de Nyquist, la frecuencia observada en la señal discreta se pliega dentro del rango [0,fm/2]. Es decir que señales con frecuencias que difieren en multiplos enteros de la frecuencia de muestreo fm generan exactamente las mismas muestras, por ello, la señal que se produce en Nyquist  (ff = 50 Hz) es identica a la señal que se produce en ff = 500 Hz ( 10*50 = 500), de la misma manera ocurre con las señales ff = 2001 Hz y 1001 Hz, ya que graficamente son señales de ff = 1Hz. Para el caso de frecuencias de señales en el rango [fm/2 , 99], las señales se veran desfasadas, por eso es que en ff = 999 hz se ve como en ff = 99 hz (una señal senoidal invertida).


### *Implementacion de señal cuadrada*

Genero una señal cuadrada con un duty de 0.5

```python
from scipy import signal
#tiene por defecto el duty de 0.5

t_cuadrada = np.arange(nn)/fm
frecuencia_cuadrada = 1
señal_cuadrada = signal.square(2*np.pi*frecuencia_cuadrada*t_cuadrada, duty = 0.5)
plt.plot(t_cuadrada, señal_cuadrada)
plt.title("Señal cuadrada")
plt.grid()
```
